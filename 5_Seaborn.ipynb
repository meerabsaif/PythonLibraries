# Data Visualization Using Seaborn
Seaborn divides the graphs in 3 categories :
>1. Relational Plots.
>
>2. Distributional Plots.
>
>3. Categorical Plots.
>
We can plot the graphs independently 

## Importing libraries 

```python
import seaborn as sns
import matplotlib as plt
```

### Applying the default theme 

```python
sns.set_theme()
```

```python
# Loading the dataset
tip = sns.load_dataset("tips")
tip.head()
```

```python
#visualize the data 
sns.relplot(data=tip,x="total_bill",y="tip")           #relplot draws scatterplot by default 
```

```python
sns.relplot(data= tip, x= "total_bill", y= "tip" , col = "sex") #col is used to split the data  #it seperates the data on the base of gender 
```

```python
sns.relplot(data= tip, x= "total_bill", y= "tip" , col = "sex", hue= "smoker")  #defining colours on the basis of smoker or non smoker 
```

```python
sns.relplot(data= tip, x= "total_bill", y= "tip" , col = "sex", hue="smoker", size= "size")      # size the no. of people
```

```python
dots= sns.load_dataset("dots")
sns.relplot(data=dots,x="time",y="firing_rate",kind="line")    #we defined the kind of plot. which is line plot. 
```

```python
dots
```

```python
sns.relplot(data = dots , x= "time", y="firing_rate", kind="line", col="align")
```

```python
sns.relplot(data = dots , x= "time", y="firing_rate", kind="line",
            col= "align",
            hue= "choice",size= "coherence"
           )

#col splits the graph 
#hue defines colour on the basis of choice
#size is determined on the basis of coherence 
```

```python
sns.relplot(data = dots , x= "time", y="firing_rate", kind="line",
            col= "align",
            hue= "choice",size= "coherence",
            style= "choice"
           )

#style assigns a style to lines. some are continous here and others are doted.
```

## Statistical Estimation 
### Using data visualization 

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data= fmri, x= "timepoint", y= "signal" , kind= "line",
    hue= "event", style ="event", col= "region"
)

#we are checking the average on the basis of any variable. here we used "region"
```

```python
df= sns.load_dataset("titanic") 
df.head() 
```

```python
sns.displot(data= df, x= "age", col="alive")

#we check distribution of data using displot 
```

```python
sns.displot(data= df, x= "age", col="alive", kde= True)
#line on data makes the distribution obvious 
```

```python
sns.displot(data= df, x= "age", col="alive", hue= "sex", kind= "ecdf")
```

```python
sns.catplot(data= df, x= "class", kind= "swarm", y="age", hue="alive")
```
