```python
import numpy as np
import matplotlib.pyplot as plt
 
  
# creating the dataset
data = {'Study':15, "Repeated Study":30, "Concept Mapping":25,
        "Retrievel Practice" :50}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='maroon',
        width = 0.4)
 
plt.xlabel("Groups")
plt.ylabel("Propotions Correct")
plt.title("Active Recall Theory Evidence")
plt.show()
```

**Scatter Chart**

>Scatter Chart is used to represent a cluster to see the range of the variables.
>
>They are used to plot data points on horizontal and vertical axis in the attempt to show how much one variable is affected by another.
>
>Each row in the data table is represented by a marker the position depends on its values in the columns set on the X and Y axes.
>
>A third variable can be set to correspond to the color or size of the markers, thus adding yet another dimension to the plot.

You can use the scatter() function to draw a scatter plot.


```python
import matplotlib.pyplot as plt

import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])

y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)

plt.show()
```

 **Histogram**

>Histogram is used as a count plot. It is used to detect how many values lie in a given range.
>
>It is an accurate representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable. 
>
>The parameters required to construct a histogram are x, bins, range, density, cimulative, histtype.

An example of histogram:

```python
from matplotlib import pyplot as plt
import numpy as np
 
 
# Creating dataset
a = np.array([22, 87, 5, 43, 56,
              73, 55, 54, 11,
              20, 51, 5, 79, 31,
              27])
 
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(a, bins = [0, 25, 50, 75, 100])
 
# Show plot
plt.show()
```

## Plotting

```python
#creating dataset

np.random.seed(187970701)                             #seed generates numbers randomly
data1 = { 
    'a' : np.arange(50),
    'b' : np.random.randint(0, 50, 50),               #gives random values from 0 to 50 
    'c' : np.random.randn (50) 
}
data1['d'] = data1['a'] + 10 *np.random.randn(50)     #adding another colum d into the above data set
                                                      # value of d is , add 10 in the value of a and multiply with a random number
data1['b'] = np.abs(data1['b']) *100

# plotting 

fig, ax = plt.subplots(figsize = (5, 2.5) , layout = 'constrained' )    #figure size = size of white bachground
ax.scatter(x = 'a', y= 'd', data=data1 )                                #a uploaded on x axis and d uploaded on y axis
                                                                  
```

```python
#creating dataset or importing. 

np.random.seed(187970701)                 #seed generates numbers randomly
data1 = { 
    'a' : np.arange(50),
    'b' : np.random.randint(0, 50, 50),   #gives random values from 0 to 50 
    'c' : np.random.randn (50) 
}
data1['d'] = data1['a'] + 10 *np.random.randn(50)     #adding another colum d into the above data set
                                                      # value of d is , add 10 in the value of a and multiply with a random number

data1['b'] = np.abs(data1['b']) *100

# plotting 
fig, ax = plt.subplots(figsize = (5, 2.5) , layout = 'constrained' )    #figure size = size of white bachground
ax.scatter(x = 'a', y= 'd', c = 'b' , s= 'd' , data=data1 )   

# a uploaded on x axis and d uploaded on y axis 
# c will be the color of dots. it will be according to the value of b.
#size will be according to the size of a""

#ADDING LABELS

ax.set_xlabel('values of a ')
ax.set_ylabel('values of b ' )
ax.set_title( ' Comparison of a and b ')
```

### Coding style :

1. Explicit
2. Implicit 

## Explicit style 

```python

x = np.linspace (0,2,200)          #data 

fig,ax = plt.subplots(figsize = (5, 2.5), layout = 'constrained' )
ax.plot( x , x , label= 'linear' )  #plot some data on axes 
ax.plot (x, x**2, label = '2nd line') #2nd data on a
ax.plot(x, x**3, label = '3rd line') 
ax.set_xlabel('x')
ax.set_ylabel('x and multiply')
ax.set_title("line plot") 
ax.legend()
```

## Implicit Style 

We use the library directly : 

```python
x = np.linspace (0,2,200)          #data 

plt.subplots(figsize = (5, 2.5), layout = 'constrained' )
plt.plot( x , x , label= 'linear' )  #plot some data on axes 
plt.plot (x, x**2, label = '2nd line') #2nd data on a
plt.plot(x, x**3, label = '3rd line') 
plt.xlabel('x')
plt.ylabel('x and multiply')
plt.title("line plot") 

#we also also give limitations, where it starts and ends 
plt.xlim(0 , 7 )                       #from 0 to 10
plt.ylim(0 , 3 )
plt.legend()
```
